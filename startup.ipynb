```python
cities = df['City  Location']
```

```python
cities.unique()
```

```python
cities.fillna('None', inplace = True)
```

```python
print(cities.unique())
```

```python
cities = list(cities)

final = {}

for city in cities:
    
    if city == 'bangalore':
        city = 'Bangalore'
    if city == "Delhi":
        city = 'New Delhi'
        
    if '/' in city:
        key = city.split('/')  # ['city1', 'city2']
        if key[0] != '':
            final[key[0].strip()] = final.get(key[0].strip(), 0) + 1
            
    else:
        if city != 'None':
            final[city] = final.get(city, 0) + 1
    
final = dict(sorted(final.items(), key= lambda item: item[1], reverse = True))

count = 0
for key, value in final.items():
    if count == 10:
        break
    count += 1
    print(key, value)
```

```python
plt.pie(list(final.values())[:10], labels = list(final.keys())[:10])
plt.show()
```
